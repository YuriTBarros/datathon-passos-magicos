#### Estratégia de Padronização e Unificação de Dados (Master Schema Completo)

##### 1. Diretrizes de Governança e Transformação
Para garantir que o modelo de Machine Learning opere sobre um fundamento sólido, aplicaremos as seguintes transformações estruturais:

* **Padronização Sintática:** Todas as colunas em `snake_case` (letras minúsculas e sublinhados). Ex: `Gênero` ➔ `genero`.
* **Anonimização de Identidade:** A coluna será `nome`, mas o conteúdo será o dado anonimizado (`Aluno-XXX`).
* **Normalização Temporal:**
    * `ano_nascimento`: Extração apenas do ano (4 dígitos) para uniformidade.
    * `ano`: Nova coluna obrigatória para identificar o registro (2022, 2023 ou 2024).
* **Tratamento de Status (Atividade):** Nova coluna `situacao_ativo`. Premissa: `Se INDE > 0 ➔ Ativo`. Em 2024, será feita a validação cruzada com a coluna original.
* **Agregação de Listas:**
    * `avaliadores`: Consolidação de `Avaliador1` até `Avaliador6` em uma estrutura de lista.
    * `recomendacoes`: Consolidação de `Rec Av1` até `Rec Av4` em uma estrutura de lista.
* **Limpeza de Escopo:** Exclusão da coluna `escola` por inconsistência histórica entre os datasets.

#### Tratamento do IPP de 2022

Para garantir a continuidade dos dados e o treinamento do modelo de Machine Learning, propõe-se a recuperação do **Indicador Psicopedagógico (IPP)** via cálculo reverso.

##### 1. Metodologia de Cálculo (Baseado no documento fornecido 'PEDE: Pontos importantes'):
O INDE é uma média ponderada composta por 7 indicadores. A fórmula de composição é:

$$INDE = (IAN \cdot 0,1) + (IDA \cdot 0,2) + (IEG \cdot 0,2) + (IAA \cdot 0,1) + (IPS \cdot 0,2) + (IPP \cdot 0,1) + (IPV \cdot 0,1)$$

Para isolar o **IPP**, aplicamos a seguinte álgebra:
1. Calculamos a soma ponderada de todos os indicadores conhecidos ($S_{conhecida}$).
2. Subtraímos essa soma do valor total do $INDE$.
3. Dividimos o resultado pelo peso correspondente ao IPP ($0,1$).

**Equação de Isolamento:**
$$IPP = \frac{INDE - [(IAN \cdot 0,1) + (IDA \cdot 0,2) + (IEG \cdot 0,2) + (IAA \cdot 0,1) + (IPS \cdot 0,2) + (IPV \cdot 0,1)]}{0,1}$$

##### 2. Tratamento para Fase 8 (Universitários)
Com base nas diretrizes do projeto:
* **Regra:** Para alunos da **Fase 8**, o IPP não possui impacto no cálculo do INDE.
* **Ação:** Atribuiremos o valor fixo de **0** para o IPP nestes casos.
* **Justificativa:** Neutralização do peso acadêmico/pedagógico para o público universitário conforme as regras de negócio vigentes em 2022.

##### 4. Premissas de Validação (QA)
* O valor de IPP deve estar obrigatoriamente no intervalo $[0, 10]$.
* Caso o cálculo resulte em valores fora deste intervalo, será realizada uma auditoria nos pesos para verificar se houve alteração na fórmula específica para determinadas faixas etárias naquele ano.

```python
def recuperar_ipp_2022(row):
    # Regra de negócio: Fase 8 não utiliza IPP
    # Usamos str() para garantir a comparação independente do tipo lido
    if str(row['Fase']).strip() == '8':
        return 0.0
    
    try:
        # Extração dos valores para o cálculo
        # Nota: Usamos nomes de colunas conforme o seu print anterior
        inde = float(row['INDE 22'])
        ian  = float(row['IAN'])
        ida  = float(row['IDA'])
        ieg  = float(row['IEG'])
        iaa  = float(row['IAA'])
        ips  = float(row['IPS'])
        ipv  = float(row['IPV'])
        
        # Cálculo da soma dos componentes conhecidos com seus respectivos pesos
        soma_conhecida = (ian * 0.1) + (ida * 0.2) + (ieg * 0.2) + (iaa * 0.1) + (ips * 0.2) + (ipv * 0.1)
        
        # Isolando o IPP: (INDE - Soma) / 0.1
        ipp_calc = (inde - soma_conhecida) / 0.1
        
        # Limpeza: Arredondamento e clipping entre 0 e 10 para evitar ruído de ponto flutuante
        return round(max(0, min(10, ipp_calc)), 3)
    
    except (ValueError, TypeError):
        # Caso algum valor base seja nulo ou inválido, retorna NaN
        return np.nan

# 2. Aplicando a lógica
df_2022['IPP'] = df_2022.apply(recuperar_ipp_2022, axis=1)

# 3. Sumário para sua Análise
print("=== RESULTADOS DA RECUPERAÇÃO DO IPP (2022) ===")
print(f"Total de alunos processados: {len(df_2022)}")
print(f"IPP calculado (não nulos): {df_2022['IPP'].count()}")
print(f"Média do IPP: {df_2022['IPP'].mean():.2f}")
print(f"Mediana do IPP: {df_2022['IPP'].median():.2f}")

# 4. Verificação de Sanidade: Amostra dos 10 primeiros
print("\n--- Amostra Comparativa (Primeiros 10 registros) ---")
print(df_2022[['RA', 'Fase', 'INDE 22', 'IPP']].head(10))

# 5. Verificando a consistência da Fase 8
fase_8_check = df_2022[df_2022['Fase'].astype(str).str.contains('8')]
if not fase_8_check.empty:
    print("\n--- Verificação Fase 8 (Deve ser 0.0) ---")
    print(fase_8_check[['RA', 'Fase', 'IPP']].head())
```

#### Padronização da estrutura

Com o objetivo de acelerar o processo construtivo após EDA, para essa parte estaremos criando funções que poderão servir de base para o documento de processamento que teremos no projeto.

##### Funções Auxiliares para a Padronização

```python
def to_numeric_safe(value):
    """Converte valores para float, tratando vírgulas e strings."""
    if pd.isna(value): return np.nan
    if isinstance(value, (int, float)): return float(value)
    try:
        clean_val = str(value).replace(',', '.').strip()
        return float(clean_val)
    except:
        return np.nan

def extract_year(value):
    """Extrai o ano de forma robusta, priorizando anos isolados."""
    if pd.isna(value) or str(value).strip() == '':
        return np.nan
    
    # 1. Tenta tratar como ano direto (ex: 2005 ou "2005")
    try:
        # Se for um número grande (ex: 2005.0), float() e int() resolvem
        val_numeric = int(float(value))
        # Verifica se está em um range plausível de ano de nascimento
        if 1900 <= val_numeric <= datetime.now().year:
            return val_numeric
    except (ValueError, TypeError):
        pass

    # 2. Se falhou, tenta tratar como data completa (ex: "6/17/2015")
    try:
        return pd.to_datetime(value).year
    except:
        return np.nan

def normalize_gender(value):
    v = str(value).lower()
    if 'menina' in v or 'fem' in v: return 'feminino'
    if 'menino' in v or 'mas' in v: return 'masculino'
    return np.nan

def collect_list(row, keywords):
    cols = [c for c in row.index if any(k in c for k in keywords)]
    return [row[c] for c in cols if pd.notnull(row[c]) and str(row[c]).strip() != '']

def calculate_ipp_2022(row):
    """Engenharia reversa do IPP para 2022."""
    if str(row.get('Fase')) == '8': return 0.0
    try:
        # Usamos to_numeric_safe para garantir que os valores sejam floats
        inde = to_numeric_safe(row.get('INDE 22', 0))
        ian  = to_numeric_safe(row.get('IAN', 0))
        ida  = to_numeric_safe(row.get('IDA', 0))
        ieg  = to_numeric_safe(row.get('IEG', 0))
        iaa  = to_numeric_safe(row.get('IAA', 0))
        ips  = to_numeric_safe(row.get('IPS', 0))
        ipv  = to_numeric_safe(row.get('IPV', 0))
        
        soma = (ian*0.1) + (ida*0.2) + (ieg*0.2) + (iaa*0.1) + (ips*0.2) + (ipv*0.1)
        return round(max(0, min(10, (inde - soma) / 0.1)), 3)
    except:
        return np.nan
```

##### Aplicação de Mapeamento

```python
def unificar_safra(df_raw, mapping, ano):
    df_temp = df_raw.copy()
    
    # QA Específico por Ano
    if ano == 2022: 
        df_temp['IPP'] = df_temp.apply(calculate_ipp_2022, axis=1)
    
    if ano == 2023:
        if 'Destaque IPV.1' in df_temp.columns:
            df_temp['Destaque IPV'] = df_temp['Destaque IPV'].fillna('').astype(str) + " " + df_temp['Destaque IPV.1'].fillna('').astype(str)
            df_temp['Destaque IPV'] = df_temp['Destaque IPV'].str.strip()
    
    # 1. Rename para o Master Schema
    df_mapped = df_temp.rename(columns=mapping)
    
    # 2. Conversão Numérica Crítica (evita o erro de TypeError >)
    cols_numericas = ['inde', 'mat', 'por', 'ing', 'iaa', 'ieg', 'ips', 'ipp', 'ipv', 'ian', 'idade', 'defasagem']
    for col in cols_numericas:
        if col in df_mapped.columns:
            df_mapped[col] = df_mapped[col].apply(to_numeric_safe)
    
    # 3. Normalizações (Identidade e Tempo)
    df_mapped['ano'] = ano
    df_mapped['genero'] = df_mapped['genero'].apply(normalize_gender)
    df_mapped['ano_nascimento'] = df_mapped['nasc_raw'].apply(extract_year)
    
    # 4. Agrupamento de Listas
    df_mapped['avaliadores'] = df_temp.apply(lambda r: collect_list(r, ['Avaliador']), axis=1)
    df_mapped['recomendacoes'] = df_temp.apply(lambda r: collect_list(r, ['Rec Av']), axis=1)
    
    # 5. Lógica de Atividade (inde agora é float)
    df_mapped['situacao_ativo'] = df_mapped['inde'].apply(
        lambda x: 'Ativo' if pd.notnull(x) and x > 0 else 'Inativo'
    )
    
    # Seleção Final (Target Schema)
    target_columns = [
        'ra', 'nome', 'ano', 'ano_nascimento', 'idade', 'genero', 'ano_ingresso', 'tipo_instituicao',
        'fase', 'turma', 'inde', 'pedra', 'cg', 'cf', 'ct', 'n_av', 'iaa', 'ieg', 'ips', 'ipp',
        'ida', 'mat', 'por', 'ing', 'indicado', 'atingiu_pv', 'ipv', 'ian', 'fase_ideal', 
        'defasagem', 'rec_psicologica', 'destaque_ieg', 'destaque_ida', 'destaque_ipv',
        'situacao_ativo', 'avaliadores', 'recomendacoes'
    ]
    
    # Garantir que todas as colunas existam
    for col in target_columns:
        if col not in df_mapped.columns:
            df_mapped[col] = np.nan
            
    return df_mapped[target_columns]

```

#### Verificando indicadores de Matemática, Português e Inglês

Como os indicadores Matemática(mat),Português(por), Inglês(ing) são utilizados para calcular o **IDA** por meio de média aritmética,
para os registros onde o **IDA** está presente, mas uma das notas componentes está ausente, podemos utilizar a **Engenharia Reversa da Média**. 

Este procedimento evita que o modelo de ML trate o aluno como "nulo" em disciplinas específicas, preservando a riqueza do perfil acadêmico.

Portanto, precisamos identificar as situações que se enquandram nesse cenário.

```python
# 1. Definindo as colunas de notas acadêmicas
notas_cols = ['mat', 'por', 'ing']

# 2. Calculando a quantidade de notas faltantes por linha
df_master['n_notas_nulas'] = df_master[notas_cols].isnull().sum(axis=1)

# 3. Criando um resumo do cenário de notas
# Cruzamos o número de notas faltantes com a existência ou não do IDA
resumo_notas = df_master.groupby(['ano', 'n_notas_nulas']).agg({
    'ra': 'count',
    'ida': lambda x: x.isnull().sum()
}).rename(columns={'ra': 'total_registros', 'ida': 'ida_faltante'})

print("--- 📋 Diagnóstico de Lacunas Acadêmicas (Mat/Por/Ing) ---")
print(resumo_notas)

# 4. Identificando os casos críticos (Mais de 1 nota faltante)
mask_critico_notas = df_master['n_notas_nulas'] > 1
df_notas_criticas = df_master[mask_critico_notas].copy()

print(f"\nTotal de registros com MAIS DE UMA nota faltando: {len(df_notas_criticas)}")

# 5. Visualização detalhada dos casos críticos (amostra por fase)
colunas_view = ['ra', 'ano', 'fase', 'turma', 'mat', 'por', 'ing', 'ida']
print("\n--- 🚩 Amostra de Registros com Alta Nulidade Acadêmica ---")
display(df_notas_criticas[colunas_view].head(20))
```

Isso nos mostra que dos 6 dados que na teoria não seriam tratados, podem receber 0, já que todos possuem um ida, mesmo que para 2 deles, seja 0.

Quantos aos demais, podemos utilizar a estratégia de calculo de média. 

```python
def identificar_coluna_nula(row, colunas):
    """Retorna o nome da primeira coluna nula encontrada em uma linha."""
    for col in colunas:
        if pd.isnull(row[col]):
            return col
    return None

def calcular_nota_reversa(row):
    """
    Aplica a lógica de recomposição de Mat/Por/Ing baseada no IDA.
    Pode ser aplicada via .apply(axis=1) apenas em registros com nulos.
    """
    notas = {'mat': row['mat'], 'por': row['por'], 'ing': row['ing']}
    ida_alvo = row['ida']
    
    # Identifica nulos e preenchidos
    nulos = [k for k, v in notas.items() if pd.isnull(v)]
    preenchidos = [v for k, v in notas.items() if pd.notnull(v)]
    
    # Caso 0: Se o IDA é 0, todos os nulos acadêmicos viram 0
    if ida_alvo == 0:
        for k in nulos: notas[k] = 0.0
        return pd.Series(notas)
    
    # Caso 1: Falta exatamente 1 nota
    if len(nulos) == 1:
        # Nota_Faltante = (IDA * 3) - Soma_conhecida
        # Usamos 3 como divisor padrão das notas componentes do IDA
        valor_faltante = (ida_alvo * 3) - sum(preenchidos)
        notas[nulos[0]] = max(0.0, round(valor_faltante, 2))
        
    # Caso 2: Faltam 2 ou 3 notas
    # Se o IDA existe e é > 0, mas faltam múltiplas notas, preenchemos nulos com 0
    # para respeitar a média informada no IDA (conforme amostra analisada)
    else:
        for k in nulos:
            notas[k] = 0.0
            
    return pd.Series(notas)

def pipeline_recomposicao_academica(df):
    """
    Função principal que gerencia o fluxo de limpeza das notas.
    """
    df_temp = df.copy()
    cols_notas = ['mat', 'por', 'ing']
    
    # Filtro: Tem IDA mas falta pelo menos uma nota
    mask = (df_temp['ida'].notnull()) & (df_temp[cols_notas].isnull().any(axis=1))
    
    if mask.any():
        print(f"Processando recomposição para {mask.sum()} registros...")
        # Aplicamos a função modular de cálculo
        df_temp.loc[mask, cols_notas] = df_temp[mask].apply(calcular_ajuste_academico_reverso, axis=1)
    
    return df_temp

# Mapeando para o nome que usaremos no apply para manter a clareza
calcular_ajuste_academico_reverso = calcular_nota_reversa
```

```python
# Aplicando a Recomposição Acadêmica (Mat/Por/Ing)
# Chamamos a função de pipeline que gerencia o filtro e a aplicação do cálculo
df_master = pipeline_recomposicao_academica(df_master)

# Verificação imediata após aplicação
print("--- ✅ Verificação de Aplicabilidade ---")
nulos_restantes = df_master[['mat', 'por', 'ing']].isnull().sum()
print(f"Notas faltantes após o tratamento:\n{nulos_restantes}")

```

#### Tratamento de CT, CG e CF
No arquivo **Dicionário de Dados Datathon.pdf**, as siglas **CG**, **CF** e **CT** referem-se aos rankings Geral, por Fase e por Turma, respectivamente. Identificamos que esses campos apresentam alta taxa de nulos. Utilizaremos o **Recálculo Baseado em Regras** sobre o `INDE`, garantindo:
* **Completude:** Preenchimento de 100% das lacunas onde houver nota.
* **Fidelidade Pedagógica:** O ranking refletirá exatamente a nota atual do aluno após a limpeza.

##### Metodologia de Cálculo
O tratamento será realizado através de janelas de partição:
* **CG (Geral):** Ranking do `inde` por `ano`.
* **CF (Fase):** Ranking do `inde` por `ano` e `fase`.
* **CT (Turma):** Ranking do `inde` por `ano`, `fase` e `turma`.
* **Regra de Empate:** Método `min` (padrão acadêmico).

```python
def tratar_e_recalcular_rankings(df):
    """
    Trata nulos e recalcula as classificações CG, CF e CT 
    baseando-se no INDE para garantir integridade.
    """
    df_clean = df.copy()
    
    # 1. Garantir que o INDE seja numérico para o ranking funcionar
    df_clean['inde'] = pd.to_numeric(df_clean['inde'], errors='coerce')
    
    # 2. Definir quem entra no ranking (apenas alunos com INDE válido e > 0)
    # Alunos inativos ou sem nota não devem "sujar" a classificação dos outros
    mask_ativo = (df_clean['inde'].notnull()) & (df_clean['inde'] > 0)
    
    print(f"Recalculando rankings para {mask_ativo.sum()} registros ativos...")

    # 3. CG - Classificação Geral (por Ano)
    df_clean.loc[mask_ativo, 'cg'] = df_clean[mask_ativo].groupby('ano')['inde'] \
                                     .rank(ascending=False, method='min')

    # 4. CF - Classificação na Fase (por Ano e Fase)
    df_clean.loc[mask_ativo, 'cf'] = df_clean[mask_ativo].groupby(['ano', 'fase'])['inde'] \
                                     .rank(ascending=False, method='min')

    # 5. CT - Classificação na Turma (por Ano, Fase e Turma)
    df_clean.loc[mask_ativo, 'ct'] = df_clean[mask_ativo].groupby(['ano', 'fase', 'turma'])['inde'] \
                                     .rank(ascending=False, method='min')

    return df_clean

# Execução do tratamento
df_master = tratar_e_recalcular_rankings(df_master)

# Verificação visual dos resultados pós-tratamento
print("\n--- Verificação casos que não foram preenchidos devido a não ter um INDE ---")
print(df_master[['cg', 'cf', 'ct']].isnull().sum())
```

#### Checagem de idade

```python
def diagnosticar_recuperacao_idade(df):
    """
    Analisa quantos nulos de idade possuem o ano_nasc preenchido.
    """
    # Filtro de quem não tem idade
    nulos_idade = df[df['idade'].isnull()].copy()
    
    # Verifica se a coluna de ano de nascimento (ex: 'ano_nasc') tem dados
    tem_ano_nasc = nulos_idade['ano_nascimento'].notnull().sum()
    total_nulos = len(nulos_idade)
    
    print(f"--- 📊 Diagnóstico de Idade ---")
    print(f"Total de idades nulas: {total_nulos}")
    print(f"Recuperáveis via Ano de Nascimento: {tem_ano_nasc} ({ (tem_ano_nasc/total_nulos)*100:.1f}%)")
    
    return nulos_idade

def aplicar_calculo_idade(df):
    """
    Preenche a idade usando a conta: Ano de Referência - Ano de Nascimento.
    """
    df_temp = df.copy()
    
    # Máscara: Idade é nula mas o Ano de Nascimento existe
    mask = (df_temp['idade'].isnull()) & (df_temp['ano_nascimento'].notnull())
    
    # Cálculo direto
    df_temp.loc[mask, 'idade'] = df_temp.loc[mask, 'ano'] - df_temp.loc[mask, 'ano_nascimento']
    
    # Garante que a idade seja um número inteiro (opcional)
    df_temp['idade'] = df_temp['idade'].astype(float)
    
    print(f"✅ Sucesso: {mask.sum()} registros de idade foram recuperados matematicamente.")
    return df_temp
```

```python
# Primeiro, recuperamos o máximo possível com precisão (Data de Nascimento)
df_master = aplicar_calculo_idade(df_master)

# Validamos com o seu mapa de calor
plot_null_heatmap(df_master, titulo='Validação Final: Comparativo longitudinal(2022-2024)')
```

#### Análise e Cálculo de 'Atingiu PV'

De acordo com o documento **Relatório PEDE 2022**, a determinação se um aluno atingiu o ponto de virada ou não é feita com uma nota de corte do IPV.  SE  IPV do aluno >= (Média_Ano + Desvio_Padrão_Ano) ele atingiu o ponto de virada, caso seja menor, o aluno não atingiu. Dessa forma, podemos calcular para 2023 e 2024.

```python
def calcular_atingiu_pv(df):
    """
    Classifica se o aluno atingiu o Ponto de Virada (PV) baseado em:
    IPV >= (Média_Ano + Desvio_Padrão_Ano)
    """
    df_temp = df.copy()
    
    # 1. Filtramos apenas os anos de interesse (2023 e 2024)
    anos_alvo = [2023, 2024]
    
    for ano in anos_alvo:
        # Isolar os dados do ano
        mask_ano = df_temp['ano'] == ano
        ipv_ano = df_temp.loc[mask_ano, 'ipv']
        
        # 2. Calcular estatísticas do IPV para este ano
        media_ipv = ipv_ano.mean()
        std_ipv = ipv_ano.std()
        threshold = media_ipv + std_ipv
        
        print(f"--- 📈 Estatísticas IPV {ano} ---")
        print(f"Média: {media_ipv:.3f} | Desvio Padrão: {std_ipv:.3f}")
        print(f"Régua de Corte (Limite): {threshold:.3f}")
        
        # 3. Aplicar a regra de classificação
        # Onde o IPV for >= threshold -> 'Sim', senão 'Não'
        # Usamos np.where para eficiência ou uma lógica direta
        df_temp.loc[mask_ano, 'atingiu_pv'] = df_temp.loc[mask_ano, 'ipv'].apply(
            lambda x: 'Sim' if x >= threshold else 'Não'
        )
        
        # Contagem para validação
        qtd_sim = (df_temp.loc[mask_ano, 'atingiu_pv'] == 'Sim').sum()
        print(f"Resultado: {qtd_sim} alunos atingiram o PV em {ano}.\n")
        
    return df_temp
```

#### Análise e inputação de 'indicado' 

Não foi encontrado em nenhum documento uma menção de cálculo que classifica se o estudante é indicado para bolsa ou não. Encontramos somente a menção abaixo no seguinte documento: **'Relatório de Atividades de 2023'**.

*"...bolsas de estudo para aqueles que demonstram comprometimento e dedicação em seus estudos. "*

Diante disso, como maneira de preencimento dos dados, decidimos por tomar uma abordagem similar ao 'Atingiu PV', só que utilizaremos o **INDE**. 

Assumiremos que um estudante é indicado para bolsa se: INDE >= (Média_Ano + Desvio_Padrão_Ano). Para validar essa hipotése antes de realizar o preenchimento. Faremos a comparação com o ano de 2022.

```python
import pandas as pd
import numpy as np

def validar_hipotese_bolsa_2022(df):
    """
    Isola os dados de 2022 do DataFrame Master e valida se a regra 
    estatística (Média + DP) condiz com as indicações reais registradas.
    """
    # 1. Isolar dados de 2022 a partir do Master
    df_2022 = df[df['ano'] == 2022].copy()
    
    # Garantir que temos dados para calcular
    if df_2022.empty:
        print("⚠️ Erro: Não foram encontrados registros de 2022 no DataFrame fornecido.")
        return None

    # 2. Padronização da coluna de indicação real (INDICADO)
    # Convertemos para String, removemos espaços e padronizamos para 'Sim'/'Não'
    df_2022['INDICADO'] = df_2022['indicado'].astype(str).str.strip().str.capitalize()
    
    # 3. Cálculo das Métricas Estatísticas do INDE (2022)
    # Removemos nulos apenas para o cálculo da média/dp
    media_inde = df_2022['inde'].mean()
    std_inde = df_2022['inde'].std()
    threshold = media_inde + std_inde
    
    # 4. Aplicação da Regra (Predição baseada na Hipótese)
    df_2022['predicao_bolsa'] = df_2022['inde'].apply(
        lambda x: 'Sim' if x >= threshold else 'Não'
    )
    
    # --- RELATÓRIO DE RESULTADOS ---
    print(f"--- 🧪 Validação de Hipótese (Safra 2022) ---")
    print(f"Média INDE 2022: {media_inde:.3f}")
    print(f"Desvio Padrão: {std_inde:.3f}")
    print(f"Nota de Corte (Threshold): {threshold:.3f}\n")
    
    # 5. Matriz de Comparação (Real vs. Hipótese)
    cruzamento = pd.crosstab(
        df_2022['INDICADO'], 
        df_2022['predicao_bolsa'], 
        rownames=['Real (Histórico)'], 
        colnames=['Hipótese (Estatística)']
    )
    
    print("--- 📊 Matriz de Comparação ---")
    print(cruzamento)
    
    # 6. Cálculo de Acurácia
    acertos = np.diag(cruzamento).sum()
    total = cruzamento.values.sum()
    acuracia = (acertos / total) * 100
    
    print(f"\nAcurácia da Hipótese: {acuracia:.1f}%")
    
    # 7. Identificação de Falsos Negativos (Casos onde a dedicação superou a nota)
    falsos_negativos = df_2022[(df_2022['INDICADO'] == 'Sim') & (df_2022['predicao_bolsa'] == 'Não')]
    
    if not falsos_negativos.empty:
        print(f"\n💡 Insight: {len(falsos_negativos)} alunos foram indicados em 2022 mas ficaram abaixo do corte.")
        print("Isso valida que o 'comprometimento' do relatório pode ter nuances qualitativas.")
        
    return df_2022

```

```python
df_resultado_val_2022 = validar_hipotese_bolsa_2022(df_master)
```

Isso nos mostra que a hipotése está na direção correta, mas que provavelmente somente o INDE não seja o suficiente para classificar. Por isso, vamso a pena fazer um teste de sensibilidade para melhorar essa classificação. 

##### Verificação por desvio padrão

```python
def otimizar_threshold_bolsa(df):
    df_2022 = df[df['ano'] == 2022].copy()
    df_2022['real'] = df_2022['indicado'].astype(str).str.strip().str.capitalize()
    
    media = df_2022['inde'].mean()
    std = df_2022['inde'].std()
    
    # Testando diferentes multiplicadores para o Desvio Padrão
    for m in [0, 0.25, 0.5, 0.75, 1.0]:
        corte = media + (m * std)
        pred = df_2022['inde'].apply(lambda x: 'Sim' if x >= corte else 'Não')
        
        # Cálculo de acurácia simples
        acuracia = (pred == df_2022['real']).mean() * 100
        print(f"Multiplicador {m}x | Corte: {corte:.3f} | Acurácia: {acuracia:.1f}%")

# Executar para ver qual multiplicador chega mais perto do real
otimizar_threshold_bolsa(df_master)
```

Os resultados mostram que quanto mais subimos a "régua" (o corte), maior a acurácia global. Isso acontece porque a grande maioria dos alunos não é indicada para bolsa, então a classificação fica muito boa em prever o "Não". No entanto, há um problema: no multiplicador 1.0x, por exemplo, apesar da acurácia de 77%, a abordagem é muito rígido e deixa de fora muitos alunos que ganharam bolsa na vida real (os Falsos Negativos). Por isso, vamos tentar uma nova hipótese.

Vamos testar uma regra composta. O aluno é indicado se:
Tiver um INDE excelente (Top Performance), **OU** se tiver um INDE acima da média E um IEG (Engajamento) muito alto.

```python
def validar_hipotese_composta(df):
    df_2022 = df[df['ano'] == 2022].copy()
    df_2022['real'] = df_2022['indicado'].astype(str).str.strip().str.capitalize()
    
    # Parâmetros Estatísticos
    media_inde = df_2022['inde'].mean()
    std_inde = df_2022['inde'].std()
    corte_rigido = media_inde + std_inde # O seu corte de 8.054
    
    # 1. Definição da Regra Composta
    # Critério A: INDE excepcional (>= 8.05)
    # Critério B: INDE acima da média (> 7.03) E Engajamento (IEG) excelente (ex: >= 9.0)
    def aplicar_regra(row):
        if row['inde'] >= corte_rigido:
            return 'Sim'
        if (row['inde'] > media_inde) and (row['ieg'] >= 9):
            return 'Sim'
        return 'Não'

    df_2022['predicao_composta'] = df_2022.apply(aplicar_regra, axis=1)

    # 2. Matriz de Confusão
    cruzamento = pd.crosstab(df_2022['real'], df_2022['predicao_composta'], 
                             rownames=['Real'], colnames=['Hipotese Composta'])
    
    acuracia = (df_2022['real'] == df_2022['predicao_composta']).mean() * 100
    
    print(f"--- 🧪 Validação: Regra Composta (INDE + IEG) ---")
    print(f"Acurácia: {acuracia:.1f}%")
    print("\n--- 📊 Matriz de Comparação ---")
    print(cruzamento)

    
    return df_2022

# Executar a validação
df_val_composta = validar_hipotese_composta(df_master)
```

A acurácia caiu para 65.5% porque, ao adicionarmos o critério de engajamento (IEG), acabamos sendo "generosos" demais: a regra indicou 217 alunos que, na vida real, não receberam a bolsa (Falsos Positivos).

No entanto, note um ponto positivo: conseguimos "resgatar" mais bolsistas reais. O número de acertos no "Sim" subiu de 34 para 52. Isso prova que o comprometimento (IEG) faz parte da conta, mas a régua dele precisa ser mais restrita ou combinada com outros fatores.


##### Utilização de um modelo de Machine Learning para Preenchimento da coluna indicado

Devido aos resultados encontrados, e levando em consideração os seguintes pontos, decidimos por tomar a abordagem de criar um modelo simples de classificação para fazer o preenchimento de valores faltantes na coluna de indicação de bolsas:

1. Tradução da "Inteligência de Negócio"
   
A indicação para bolsas na Passos Mágicos não é um evento aleatório. O relatório de atividades é claro ao citar que as bolsas são destinadas a quem demonstra **comprometimento e dedicação**. Nossos testes estatísticos manuais mostraram que o comportamento humano e os critérios da ONG seguem padrões repetíveis que fórmulas simples de média não capturam integralmente. O modelo de Machine Learning (ML) pode ser capaz de traduzir termos subjetivos em correlações matemáticas.

2. Superação da Subjetividade com Consistência Matemática
   
Ao analisarmos os dados de 2022, observamos que apenas a nota (INDE) não explica a totalidade das bolsas concedidas. Há um "fator de mérito" ligado à evolução e ao tipo de instituição de ensino que precisava de uma análise multidimensional.


3. Identificação de "Proxies" de Sucesso para Risco de Defasagem
   
Nosso objetivo final é estimar o risco de defasagem do aluno. Ter métricas de sucesso é importantíssimo para classificar. Está certo que ser indicado para a bolsa é o resultado composto de outras métricas, mas ter ela pode auxiliar na predição.


```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

def treinar_modelo_bolsa_v3(df):
    # 1. Preparação dos dados de 2022
    df_train = df[df['ano'] == 2022].copy()
    
   
    # Criamos uma função simples de limpeza para garantir o match
    def mapear_escola(valor):
        val = str(valor).strip().upper()
        if 'PÚBLICA' in val or 'PUBLICA' in val:
            return 1
        else:
            return 0
        return 1 # Default para Pública, que é a maioria

    df_train['tipo_inst_num'] = df_train['tipo_instituicao'].apply(mapear_escola)
    
    # 2. Seleção de variáveis (incluindo as colunas de componentes cf, cg, ct)
    features = ['inde', 'ieg', 'ida', 'ipv', 'ian', 'ipp', 'ips', 'cf', 'cg', 'ct', 'tipo_inst_num']
    
    X = df_train[features].fillna(df_train[features].mean())
    y = df_train['indicado'].astype(str).str.strip().str.capitalize().map({'Sim': 1, 'Não': 0})
    
    # 3. Treino do Modelo
    rf = RandomForestClassifier(
        n_estimators=200, 
        max_depth=5, 
        max_leaf_nodes=10,
        class_weight='balanced', 
        random_state=42
    )
    rf.fit(X, y)
    
    print("--- 🤖 Performance do Modelo (Base 2022) ---")
    print(classification_report(y, rf.predict(X), target_names=['Não', 'Sim']))
    
    # Ranking de Importância
    importances = pd.DataFrame({'feature': features, 'importance': rf.feature_importances_})
    print("\n--- 📊 Ranking de Importância ---")
    print(importances.sort_values(by='importance', ascending=False))
    
    return rf, features, mapear_escola

# Executar Treino
modelo_rf, colunas_ml, funcao_map = treinar_modelo_bolsa_v3(df_master)
```



| Critério | Acurácia (2022) | Recall (Acerto de Bolsistas) | Justificativa |
| :--- | :---: | :---: | :--- |
| **Média Estatística** | 77.2% | 25.7% | Rígida demais; ignora o fator humano e social. |
| **Regra Híbrida** | 77.0% | 27.2% | Melhor fundamentada, mas ainda limitada em volume. |
| **Random Forest (ML)** | **87.0%** | **89.0%** | **Vencedora:** Captura o perfil holístico do aluno da ONG. |

```python
def preencher_coluna_indicado(df, modelo, features, func_map):
    df_result = df.copy()
    
    # 1. Preparar a feature de instituição (necessária para o modelo)
    df_result['tipo_inst_num'] = df_result['tipo_instituicao'].apply(func_map)
    
    # 2. Identificar onde precisamos preencher (valores nulos ou strings vazias)
    # Filtramos 2023 e 2024 que sabemos estarem vazios
    mask_vazio = (df_result['indicado'].isnull()) | (df_result['indicado'] == '') | (df_result['ano'].isin([2023, 2024]))
    
    if mask_vazio.any():
        # Selecionamos apenas as linhas que precisam de preenchimento
        X_imputar = df_result.loc[mask_vazio, features].copy()
        
        # Tratamento de nulos nas features para o modelo não falhar
        X_imputar = X_imputar.fillna(X_imputar.mean())
        
        # Predição (0 ou 1)
        preds = modelo.predict(X_imputar)
        
        # Converte 0/1 para Não/Sim
        preds_labels = ['Sim' if p == 1 else 'Não' for p in preds]
        
        # Grava DIRETAMENTE na coluna original
        df_result.loc[mask_vazio, 'indicado'] = preds_labels
        
        print(f"✅ Preenchimento concluído: {len(preds_labels)} valores imputados na coluna 'indicado'.")
    
    # 3. Padronização final (Garantir que tudo esteja como 'Sim' ou 'Não')
    df_result['indicado'] = df_result['indicado'].astype(str).str.strip().str.capitalize()
    
    # Removemos a coluna auxiliar numérica para limpar o DF
    if 'tipo_inst_num' in df_result.columns:
        df_result = df_result.drop(columns=['tipo_inst_num'])
        
    return df_result

# Executar a imputação direta
df_master = preencher_coluna_indicado(df_master, modelo_rf, colunas_ml, funcao_map)

# 📊 Verificação Final
print("\n--- 🔎 Conferência da Coluna 'indicado' ---")
print(df_master.groupby('ano')['indicado'].value_counts())
```

```python
import re
def normalizar_fases(valor):
    # Converte para string e limpa espaços
    v = str(valor).strip().upper()
    
    # 1. Caso Especial: ALFA vira 0
    if 'ALFA' in v:
        return 0
    
    # 2. Extrair apenas o primeiro dígito numérico (ex: '1A', 'FASE 1', '1.0' -> 1)
    match = re.search(r'\d', v)
    if match:
        return int(match.group())
    
    # 3. Fallback: Se não encontrar número nem ALFA, retorna nulo para tratar depois
    return None

# Aplicando a normalização nas duas colunas
df_master['fase_num'] = df_master['fase'].apply(normalizar_fases)
```

```python
def consolidar_instituicao(valor):
    v = str(valor).strip().upper()
    
    # 1. Grupo Pública
    if 'PÚBLICA' in v or 'PUBLICA' in v:
        return 'PÚBLICA'
    
    # 2. Grupo Privada (Parcerias, Apadrinhamento, Rede Decisão, JP II)
    if 'PRIVADA' in v or 'REDE DECISÃO' in v or 'JP II' in v:
        return 'PRIVADA'
    
    # 3. Grupo Especial (Concluído ou Outros)
    if 'CONCLUIU' in v:
        return 'CONCLUÍDO'
        
    return 'OUTROS'

# Aplicando a consolidação
df_master['tipo_inst_consolidado'] = df_master['tipo_instituicao'].apply(consolidar_instituicao)

# Agora transformamos em numérica para o modelo (Label Encoding)
# Pública costuma ter maior risco de defasagem sistêmica, então daremos um peso maior
mapa_inst = {
    'PÚBLICA': 2,
    'OUTROS': 1,
    'PRIVADA': 0,
    'CONCLUÍDO': 0
}

df_master['tipo_inst_num'] = df_master['tipo_inst_consolidado'].map(mapa_inst)
```

Agora sim estamos preparados para começar a pensar em no problema em questão: **Risco de defasagem**


Embora o **IAN (Indicador de Adequação de Nível)** seja uma métrica institucional valiosa para o diagnóstico imediato, sua natureza é puramente **determinística e cadastral**. Por ser calculado através de uma fórmula direta (Fase Efetiva vs. Fase Ideal por Idade), utilizá-lo como o alvo principal de um modelo de Machine Learning criaria uma "falsa acurácia". O modelo aprenderia apenas a replicar uma conta aritmética já conhecida, em vez de descobrir padrões ocultos de comportamento ou risco.

Para transformar o dataset em uma ferramenta de predição real e estratégica para a Passos Mágicos, decidimos abandonar a predição do estado estático (IAN) e focar em duas dimensões dinâmicas:

**A. Medição de Crescimento e Decaimento da Defasagem (Delta $\Delta$)**
Em vez de perguntar "quem está defasado", passamos a analisar a **velocidade de recuperação**. 
* **Por que faz sentido:** Um aluno com 2 anos de defasagem que reduz para 1 ano (Delta negativo) representa um sucesso pedagógico, enquanto um aluno sem defasagem que estagna (Delta zero ou positivo) representa um risco futuro. 
* **Classificação:** Esta abordagem permite medir o **Momentum** do aluno, transformando a defasagem em uma variável de fluxo, identificando quem está ganhando ou perdendo tração no aprendizado.

**B. Predição do Risco de Evasão (Churn)**
A evasão é o risco final da vulnerabilidade social e acadêmica. Criamos um alvo preditivo baseado na continuidade do aluno entre anos(2022-2024).
* **Por que faz sentido:** Identificar o risco de abandono antes que ele ocorra(entre um ano e o outro), permite que a associação realize buscas ativas e intervenções psicossociais preventivas. 
* **Lógica:** O modelo utiliza os indicadores comportamentais (IEG), psicossociais (IPP/IPS) e acadêmicos (IDA) do ano atual para prever a probabilidade de ausência no ano subsequente.

Ao focar no **Delta da Defasagem** e no **Risco de Evasão**, o projeto deixa de ser um relatório de "fotos do passado" para se tornar um sistema de **Alerta Antecipado (Early Warning System)**. Esta abordagem é tecnicamente superior por lidar com a causalidade dos problemas e oferecer aos gestores da Passos Mágicos a oportunidade de agir sobre o futuro, e não apenas tabular o presente.

Diante do exposto, vamos realizar os ajustes finais do dataset para fazer isso acontencer.

```python
def rotular_evasao(df):
    """
    Cria a variável target 'evadiu':
    0 = Aluno permaneceu no ano seguinte (Não evadiu)
    1 = Aluno desapareceu no ano seguinte (Evadiu)
    """
    df_temp = df.copy()
    
    # 1. Identificar todos os RAs presentes em cada ano
    ras_por_ano = {ano: set(df_temp[df_temp['ano'] == ano]['ra']) for ano in df_temp['ano'].unique()}
    
    def checar_presenca_proximo_ano(row):
        ano_atual = row['ano']
        ano_proximo = ano_atual + 1
        ra = row['ra']
        
        # Se não temos dados do ano seguinte no dataset, retornamos NaN (caso de 2024)
        if ano_proximo not in ras_por_ano:
            return np.nan
        
        # Se o RA está no set do ano seguinte, NÃO evadiu (0)
        if ra in ras_por_ano[ano_proximo]:
            return 0
        else:
            # Se sumiu, evadiu (1)
            return 1

    df_temp['evadiu'] = df_temp.apply(checar_presenca_proximo_ano, axis=1)
    
    # Nota importante: Alunos que concluíram a Fase 7 ou 8 não devem ser contados como evasão
    # se desaparecerem, pois eles "se formaram".
    mask_formados = (df_temp['evadiu'] == 1) & (df_temp['fase_num'].isin([7, 8]))
    df_temp.loc[mask_formados, 'evadiu'] = 0
    
    return df_temp
```

```python
df_master = rotular_evasao(df_master)
```

```python
# No seu notebook ou script de ML:

# 1. Base de Treino e Teste (Onde temos o "gabarito")
df_modelagem = df_master[df_master['evadiu'].notnull()].copy()

# 2. Base de Predição (Onde queremos descobrir o risco para agir agora)
df_pred_2024 = df_master[df_master['ano'] == 2024].copy()
```
